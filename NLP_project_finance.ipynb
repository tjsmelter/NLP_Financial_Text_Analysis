```python
# read TSV file into a pandas data frame
# sep="\t" tells pandas that the columns are separated by tabs (TSV format)
df = pd.read_csv("stocks.tsv", sep='\t')
```

```python
# convert the "Symbol" column of the data frame 'df' to a python list that contains stock ticker symbols
symbols = df.Symbol.tolist()

# convert the "CompanyName" column of the data frame 'df' to a python list
# list contains full names of companies (Apple Inc.)
companies = df.CompanyName.tolist()
print(symbols[:10])
```

```python
# create an entity ruler to work in the different symbols and company names from the file
nlp = spacy.blank("en")
ruler = nlp.add_pipe("entity_ruler")
patterns = []
for symbol in symbols:
  patterns.append({"label": "STOCK","pattern": symbol})
for company in companies:
  patterns.append({"label": "COMPANY", "pattern": company})
ruler.add_patterns(patterns)
```

```python
# read TSV file into a pandas data frame
# sep="\t" tells pandas that the columns are separated by tabs (TSV format)
df3 = pd.read_csv("stock_exchanges.tsv", sep="\t")
df3
```

```python
# combine three columns ('ISOMIC', 'Google Prefix', and 'Description') from the data fram "df3" into a single list called 'exchanges'
# concatenate using + operator
exchanges = df3.ISOMIC.tolist()+df3["Google Prefix"].tolist()+df3.Description.tolist()

# print first 10 items in combined exchanges
print(exchanges[:10])
```

```python
# add a list of "stops" up above for things that should not be included in the model. for example "two company"
# add ability to capture the the missing indexes and stocks

# list of words to skip when adding company names to th patterns
stops = ["two"]

# create blank English spaCy pipeline
nlp = spacy.blank("en")

# for rule-based entity recognition, add EntityRuler component to pipeline
ruler = nlp.add_pipe("entity_ruler")

# letters used for recognizing stock symbols
letters = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

# initialize empty list to hold all pattern rules
patterns = []

# add patterns for stock symbols
for symbol in symbols:
  patterns.append({"label": "STOCK","pattern": symbol})
  for l in letters:
    patterns.append({"label": "STOCK", "pattern": symbol+f".{l}"})

# add patterns for company names (while skipping stops)
for company in companies:
  if company not in stops:
    patterns.append({"label": "COMPANY", "pattern": company})

# add patterns for indexes (e.g., "S&P 500", "Nasdaq Composite")
for index in indexes:
  patterns.append({"label": "INDEX","pattern": index})
  # Add a pattern for just the first two words of the index name
  # to help catch partial mentions like "S&P"
  words = index.split()
  patterns.append({"label": "INDEX","pattern": " ".join(words[:2])})

# add patterns for index symbols (e.g., "^GSPC")
for index in index_symbols:
  patterns.append({"label": "INDEX","pattern": index})

# add the exchanges to the model (e.g., "NASDAQ", "NYSE")
for e in exchanges:
  patterns.append({"label": "STOCK_EXCHANGE","pattern": e})
ruler.add_patterns(patterns)
```

```python
# process the input text using spaCy NLP pipeline
# applies all components (like EntityRuler) and returns a doc object
doc = nlp(text)

# loop through all named entities recognized in processed document
for ent in doc.ents:
  # print entity's text (the exact words from the input) and its label (like STOCK,COMPANY, etc)
  print(ent.text, ent.label_)
# displacy.render(doc,style="ent")
```
